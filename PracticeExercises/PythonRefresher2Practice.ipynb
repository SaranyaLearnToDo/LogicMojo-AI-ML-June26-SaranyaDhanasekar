Write a program to check a palindrome.

```python
text = str(input())
if text.lower() == text[::-1].lower():
  print("Given Input String is a Palindrome")
else:
  print("Given Input String is Not a Palindrome")
```


  Write program count the blank spaces.

```python
text = str(input())
count = 0
for i in text:
  if i == " ":
    count += 1
print(f"count= {count}")
```

Write a program to Reversing a String without using reverse().

```python
text = str(input())
reverse = ""
for i in text:
  reverse = i + reverse
print(f"Reverse of {text} is {reverse}")
```
